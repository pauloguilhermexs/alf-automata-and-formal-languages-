**Objetivo:**
Desenvolver uma aplicação com a biblioteca de expressões regulares (RegEx) do python

```python
import re


# ddd dos 26 estados do brasil mais o distrito federal

ddd = {
    "84": "Rio Grande do Norte",
    "83": "Paraíba",
    "85": "Ceará",
    "71": "Bahia",
    "68": "Acre",
    "82": "Alagoas",
    "96": "Amapá",
    "92": "Amazonas",
    "61": "Distrito Federal",
    "27": "Espírito Santo",
    "62": "Goiás",
    "98": "Maranhão",
    "65": "Mato Grosso",
    "67": "Mato Grosso do Sul",
    "31": "Minas Gerais",
    "91": "Pará",
    "41": "Paraná",
    "81": "Pernambuco",
    "86": "Piauí",
    "21": "Rio de Janeiro",
    "51": "Rio Grande do Sul",
    "69": "Rondônia",
    "95": "Roraima",
    "47": "Santa Catarina",
    "11": "São Paulo",
    "79": "Sergipe",
    "63": "Tocantins"
}


entrada = input("Digite um telefone (ex: 84-66666666): ")

match = re.search(r'(\d{2})-(\d{8})', entrada) # captura o ddd-numero

if match:
    ddd_key = match.group(1) #ddd
    numero = match.group(2) # numero

    print("DDD:", ddd_key)
    print("Número:", numero)

    if ddd_key in ddd:
        print("Estado:", ddd[ddd_key])
    else:
        print("DDD não encontrado.")
else:
    print("Formato inválido.")
```
