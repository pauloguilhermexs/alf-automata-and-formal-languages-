Implementação de um programa que modela o seguinte problema: a entrada é um número inteiro positivo entre 0 (**zero**) e 999999 (**novecentos e noventa e nove mil, novecentos e noventa de nove**) e a saída fornecida é o seu valor escrito por extenso. Para outros valores, o programa informa que a entrada não é válida.

Exemplos:


*   Entrada: 204328 -> Saída: **duzentos e quatro mil, trezentos e vinte e oito**
*   Entrada: 11 -> Saída: **onze**
* Entrada: -215 -> Saída: **entrada inválida (só respondo se a entrada for um número entre 0 e 999999)**
* Entrada: casa -> Saída: **entrada inválida (só respondo se a entrada for um número entre 0 e 999999)**


Observações:



*   Utilizando o conceito de máquinas de estados finitos (**máquinas de Mealy**) para implementar solução
*   Estendendo o programa para ele responder em português (**pt**) e inglês (**en**)





```python
# funcao para tratar os casos de centenas(em portugues)
def ate_3_digitos(n):

  # em pt
  unidades = ["zero", "um", "dois", "três", "quatro", "cinco", "seis", "sete", "oito", "nove"]
  especiais = ["dez", "onze", "doze", "treze", "quatorze", "quinze", "dezesseis", "dezessete", "dezoito", "dezenove"]
  dezenas = ["", "", "vinte", "trinta", "quarenta","cinquenta", "sessenta", "setenta", "oitenta", "noventa"]
  centenas = ["", "cento", "duzentos", "trezentos", "quatrocentos","quinhentos", "seiscentos", "setecentos", "oitocentos", "novecentos"]


  c = n // 100
  d = (n % 100) // 10
  u = n % 10

  partes = []

  # maquina de estados finitos(mealy)

   # estado: centenas
  if n == 100:
    return "cem"
  elif c > 0:
    partes.append(centenas[c])

  if d == 1:
    partes.append(especiais[u])
  else:
    # estado: dezena
    if d > 1:
      partes.append(dezenas[d])
    # estado: unidade
    if u > 0:
      partes.append(unidades[u])

  return " e ".join(partes)


# funcao para tratar os casos de centenas(em ingles)

def three_digits(n):
  # in eng
  units =  ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
  specials = ["ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen", "seventeen", "eighteen", "nineteen"]
  tens =  ["", "", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]

  c = n // 100
  d = (n % 100) // 10
  u = n % 10

  partes = []

  # maquina de estados finitos(mealy)
  # estado: centenas
  if c > 0:
    partes.append(units[c] + " hundred")

  if d == 1:
    partes.append(specials[u])
  else:
    # estado: dezenas
    if d > 1:
      partes.append(tens[d])
    # estado: unidade
    if u > 0:
      partes.append(units[u])

  return " ".join(partes)


def number_in_english(n):
  milhar = n // 1000
  resto = n % 1000
  partes = []


  # maquina de estados finitos(mealy)
  #Estado: milhar
  if milhar > 0:
    partes.append(three_digits(milhar) + " thousand")

  # Estado: centenas
  if resto > 0:
    partes.append(three_digits(resto))

  return " ".join(partes)


def numero_por_extenso_pt(n):

  milhar = n // 1000
  resto = n % 1000

  partes = []

  #Maquina de estados finitos(mealy)
  #estado: milhar
  if milhar > 0:
    if milhar == 1:
      partes.append("mil")
    else:
      partes.append(ate_3_digitos(milhar) + " mil")


  #estado: centenas
  if resto > 0:
    if milhar > 0 and resto < 100:
      partes.append("e " + ate_3_digitos(resto))
    else:
      partes.append(ate_3_digitos(resto))

  return ", ".join(partes).replace(", e", " e")


# entradas e saidas

entrada = input("Digite um número: ")

try:
    n = int(entrada)

    if 0 <= n <= 999999:
        print("Português:", numero_por_extenso_pt(n))
        print("English:", number_in_english(n))
    else:
        print("entrada inválida (só respondo se a entrada for um número entre 0 e 999999)")

except:
    print("entrada inválida (só respondo se a entrada for um número entre 0 e 999999)")
```
